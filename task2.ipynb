```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Website 1
website1_url = "https://www.mygov.in/corona-data/covid19-statewise-status"
response1 = requests.get(website1_url)
soup1 = BeautifulSoup(response1.content, 'html.parser')
# Extract required data from soup1
```

```python
# Create empty lists to store the data
states = []
active_cases = []
deaths = []
recovered = []
```

```python
state_data1 = soup1.find_all('div', class_='field field-name-field-select-state field-type-list-text field-label-above')
for i in state_data1:
    states.append(i.text)
states
```

```python
active_cases_data1 = soup1.find_all('div', class_="field field-name-field-total-confirmed-indians field-type-number-integer field-label-above")
for i in active_cases_data1:
    active_cases.append(i.text)
active_cases
```

```python
deaths_data1 = soup1.find_all('div', class_='field field-name-field-deaths field-type-number-integer field-label-above')
for i in deaths_data1:
    deaths.append(i.text)
deaths
```

```python
recovered_data1 = soup1.find_all('div', class_='field field-name-field-cured field-type-number-integer field-label-above')
for i in recovered_data1:
    recovered.append(i.text)
recovered
```

```python


# Create a dictionary with the data
data = {
    'State': states,
    'Active Cases': active_cases,
    'Recovered': recovered,
    'Deaths': deaths}
```

```python
corona = pd.DataFrame(data)
corona
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(36,36))
sns.countplot(x='Deaths', data=corona)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Recovered', data=corona)
plt.xticks(rotation='vertical')
plt.show()
```
